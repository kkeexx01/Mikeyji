```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# โหลดโมเดล BERT สำหรับการจำแนกเจตนา
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')

# ฟังก์ชันการจำแนกเจตนา
def classify_intent(user_message):
    inputs = tokenizer(user_message, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    logits = outputs.logits
    prediction = torch.argmax(logits, dim=1)
    return prediction.item()

# ตัวอย่างการใช้งาน BERT
intent = classify_intent("ฉันต้องการสั่งทำโปสเตอร์")
print("เจตนาที่คาดการณ์:", intent)
```

```python
from flask import Flask, request, jsonify
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from pyngrok import ngrok

app = Flask(__name__)

# โหลดโมเดลและ tokenizer ของ GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# ฟังก์ชันในการสร้างข้อความตอบกลับ
def generate_response(prompt):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=100, do_sample=True, temperature=0.7)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

@app.route('/order', methods=['POST'])
def order():
    user_message = request.json.get("message")

    # ใช้ GPT-2 สร้างข้อความตอบกลับ
    response = generate_response(user_message)

    return jsonify({"response": response})

# เปิด ngrok tunnel
port = 5000
public_url = ngrok.connect(port)
print(f"Public URL: {public_url}")

# รัน Flask server
app.run(port=port)
```
